```python
from bs4 import BeautifulSoup
from lxml import html
import requests

pages = []
sources = []
page_count = 0
img_count = 0

#create list of pages to parse
for i in range(1, 865):
    pages.append('https://www.pokemon.com/us/pokemon-tcg/pokemon-cards/' + str(i) + '?maxhp=340')
    print(i)

print("pages all added")

#parse each page
# find cardresults div, then add src links to list
for page in pages:
    soup = BeautifulSoup(requests.get(page).text, 'html.parser')
    imgs = soup.find(id='cardResults').findAll('img')
    
    
    for img in imgs:
        sources.append(img['src'])
        img_count+=1
        print("Image count: ", img_count)
    page_count += 1
    print("PAGE COUNT: ", page_count)

print(sources)
```

```python
count = 0
with open('sources.txt', 'w') as f:
    for item in sources:
        f.write("%s\n" % item)
        count+=1
        print(count)
```
